```python
for linea in archivo.readlines():
    linea = linea.strip()
    lista = linea.split('|')
    dicc_emisiones['cod_pais'].append(lista[0])
    dicc_emisiones['nom_pais'].append(lista[1])
    dicc_emisiones['region'].append(lista[2])
    dicc_emisiones['anio'].append(lista[3])
    dicc_emisiones['co2'].append(lista[4])
    dicc_emisiones['co2_percapita'].append(lista[5])

dicc_emisiones
```

```python
for indice, elemento in enumerate(dicc_emisiones['co2']):
    elemento = elemento.replace('.','')
    elemento = elemento.replace(',','.')
    if (elemento != ''):
        elemento = float(elemento)
    else:
        elemento = 0.0
    dicc_emisiones['co2'][indice] = elemento
```

```python
# Transformación de un keys de float a int
# Cambiar el código de la siguiente línea aquí:
co2 = dicc_emisiones['co2']
for i in range(len(co2)):
    transformacionB = co2[i].replace('.', '')
    nuevo = transformacionB.replace(',', '.')
    if co2[i] != '':
        dicc_emisiones['co2'][i] = float(nuevo)
    else:
        dicc_emisiones['co2'][i] = 0.0
```

```python
# Pregunta 4
def determinar_faltantes(lista,faltante):
    conteo = 0
    for i in range(len(lista)):
        if lista[i] == faltante:
            conteo += 1
        else:
            continue
    return conteo
```
